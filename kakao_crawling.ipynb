```python
import requests
import json

rest_api_key = ""
```

```python
class KakaoLocalAPI:

  def __init__(self, rest_api_key):

    self.rest_api_key = rest_api_key
    self.headers = {'Authorization': 'KakaoAK {}'.format(rest_api_key)}
    self.URL_01 = "https://dapi.kakao.com/v2/local/search/address.json"
    self.URL_02 = "https://dapi.kakao.com/v2/local/search/category.json"
    
  def search_address(self, address):

    params = {"query": f"{address}"}
    res = requests.get(self.URL_01, headers=self.headers, params=params)
    document = json.loads(res.text)
    # return document
    match_first = document['documents'][0]['address']
    return float(match_first['x']), float(match_first['y'])
      
  def search_near_subway(self, x, y):

    params = {'category_group_code': "SW8",
              'x': f"{x}",
              'y': f"{y}",
              'radius': "1300"}

    res = requests.get(self.URL_02, headers=self.headers, params=params)
    document = json.loads(res.text)
    # return document
    match_first = document['documents'][0]
    return int(match_first['distance'])
```

```python
kakao = KakaoLocalAPI(rest_api_key)
```

```python
address = "서울특별시 도봉구 방학동 396-22"

test1 = kakao.search_address(address)
test1
```

```python
test2 = kakao.search_near_subway(test1[0], test1[1])
test2
```
